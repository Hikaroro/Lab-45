#### 1. Что такое тепловое излучение?

Тепловое излучение – это испускание электромагнитных волн за счет внутренней энергии тел. Атомы и молекулы излучающего тела переходят в возбужденное состояние за счет энергии теплового движения и при обратном переходе в невозбужденное состояние испускают электромагнитные волны.

#### 2. Что такое равновесность излучения и почему тепловое излучение, в отличие от люминесцентного, является равновесным?

Равновесное излучение – излучение, находящееся в термодинамическом равновесии с веществом. Способность теплового излучения находиться в равновесии с излучающими телами обусловлена тем, что его интенсивность возрастает при повышении температуры.

Допустим, что равновесие нарушено и тело излучает энергии больше, чем поглощает. Тогда внутренняя энергия тела будет убывать, т.е. температура тела понизится. Это приведет к уменьшению излучаемой телом энергии. Температура тела будет уменьшаться до тех пор, пока количество излучаемой телом энергии не будет равно количеству поглощаемой телом энергии. Таким образом, если равновесие нарушится, то возникают процессы, восстанавливающие равновесие.

#### 3. Почему тепловое излучение относится к квантово-оптическим явлениям? 

Тепловое излучение относится к квантово-оптическим явлениям, так как в нём проявляются квантовые свойства света.

Процесс поглощения также, как и процесс излучения электромагнитной энергии, имеем прерывистый (квантовый) характер. Особенно заметны квантовые особенности поглощения и излучения для коротких длин волн, порождаемых атомами и молекулами. Поэтому законы классической физики, полученные из наблюдений за макрообъектами, не вполне пригодны для описания процессов, происходящих на уровне атомов и молекул или еще более глубоких степенях изучения материи.

#### 4. Дайте определение основным энергетическим величинам, характеризующим тепловое излучение: поток излучения, спектральная плотность энергетической светимости, энергетическая светимость тела. Каким соотношением спектральная плотность энергетической светимости связана с энергетической светимостью тела? Что называется коэффициентами поглощения, отражения и пропускания? 

- Потоком излучения $\Phi$ называют энергию, излучаемую телом за единицу времени во всех направлениях, т.е. мощность излучения тела: $\Phi={dW \over dt}$
- Спектральной плотностью энергетической светимости называется энергия, излучаемая в единицу времени единицей поверхности тела в единичном интервале длин волн: $R_{\lambda,T}={{dE_{\lambda,\lambda+d\lambda} \over dSdtd\lambda}}$
- Энергетическая светимость – количество энергии, излучаемой единицей поверхности тела в одну секунду по всем длинам волн: $R_{T}={dE_{изл} \over dSdt}={dW \over dSdt}$, где $E_{изл}$ - полная энергия, излучаемая телом, $S$ - площадь излучающей поверхности, $t$ - время, $W$ - мощность излучения
- Связь спектральной плотности энергетической светимости с энергетической светимостью тела: $R_T=\int_{0} ^{\infty}R_{\lambda,T}d\lambda$ 
- Коэффициентом поглощения - величина, равная отношению энергии поглощенного света к энергии падающего: $\alpha={E_{\lambda погл} \over E_{\lambda пад}}$
- Коэффициентом отражения - величина, равная отношению энергии отражённого света к энергии па дающего: $\rho={E_{\lambda отр} \over E_{\lambda пад}}$
- Коэффициентом пропускания - величина, равная отношению энергии света, прошедшего через данное тело (среду) ко всей падающей энергии: $\tau={E_{\lambda прох} \over E_{\lambda пад}}$

#### 5. Какое тело называется абсолютно черным? Какова физическая модель абсолютно черного тела.

Абсолютно черное тело – это тело, которое полностью поглощает все падающее на него излучение любой частоты при любой температуре.

Физическая модель абсолютно черного тела - это очень малое отверстие в некой полости. Луч любой длины волны, попавший внутрь такой полости, может выйти из нее только после многократных отражений. При каждом отражении от стенок полости часть энергии луча поглощается и лишь ничтожная доля энергии лучей, попавших в отверстие, сможет выйти обратно; поэтому коэффициент поглощения отверстия оказывается весьма близким к единице.

#### 6. Дайте определения основным законам теплового излучения: закону Кирхгофа, закону Стефана-Больцмана, закону смещения Вина.

-  Закон Кирхгофа: отношение спектральной плотности энергетической светимости к спектральной поглощательной способности не зависит от природы тела, оно является для всех тел одной и той же (универсальной) функцией частоты (длины волны) и температуры.
- Закон Стефана-Больцмана: полная энергия, излучаемая абсолютно чёрным телом в 1с, пропорциональна четвертой степени абсолютной температуры.
- Закон смещения Вина - длина волны, на которую приходится максимум интенсивности излучения, обратно пропорциональна температуре, то есть максимум излучения с увеличением температуры смещается в сторону коротких длин волн.

#### 7. Что такое фотоны? Как гипотеза и формула Планка позволяют получить и помогают объяснить основные законы теплового излучения?

Кванты электромагнитного излучения называются фотонами.
Из формулы Планка($r_{\nu, T}={{2\pi h\nu^3}\over c^2}{1 \over e^{h\nu / kT}-1}$) можно получить частные законы теплового излучения: 
1. Если $h\nu<<kT$ (низкие частоты). $e^{h\nu / kT}\approx 1+{{h\nu} \over kT}$ $\implies$ $r_{\nu, T}={{2\pi h\nu^3}\over c^2}{kT \over h\nu}={{2\pi \nu^2}\over c^2}kT$ , т.е. получили формулу Рэлея-Джинса.
2. Если $h\nu>>kT$ (высокие частоты). В формуле Планка пренебрегаем единицей: $r_{\nu, T}={{2\pi h\nu^3}\over c^2} e^{-h\nu / kT}$ . Получили закон Вина в современной формулировке.
3. Из формулы Планка можно получить закон Стефана –Больцмана: $R_T=\int_{0} ^{\infty}r_{\nu,T}d\nu$ $=\int_{0} ^{\infty}{{2\pi h\nu^3}\over c^2}{1 \over e^{h\nu / kT}-1}d\nu$ $=[{h\nu \over kT}=x\implies \nu={kTx \over h}\implies {d\nu={kT \over h}dx}]$ $={{2\pi k^4}\over c^2 h^3}T^4\int_{0} ^{\infty}{{x^3dx}\over e^x -1}$ $=[\int_{0} ^{\infty}{{x^3 dx}\over e^x - 1}={\pi^4 \over 15}]$ = $\sigma T^4$ 
4. Формула Планка содержит закон смещения Вина. $r_{\nu, T}={{2\pi hc^2}\over \lambda^5}{1 \over e^{hc / \lambda kT}-1}$ , ${dr_{\nu, T} \over d\lambda}=0$ в максимуме. Введя обозначения $\beta={hc \over \lambda kT}$ , получим уравнение относительно $\beta$ : $5e^\beta-5-\beta e^\beta=0$ . Решение этого трансцендентного уравнения численными методами дает корень: $\beta=4,9651142$. Поэтому, $b=\lambda_m T={hc \over k\beta}=(2.897790\pm 0.000090)\times*10^{-3} м\cdot К$ .

Также измерив на опыте величины $c,\sigma,b$, можно вычислить универсальные постоянные $k$ и $h$, что и было впервые сделано Планком, причем эти значения мало отличаются от современных. 

Таким образом, формула Планка совпадает с экспериментальными данными в любой области частот, и из нее следуют все рассматриваемые нами законы и формулы. Т.е. формула Планка является полным решением теплового излучения.

#### 8. В чем заключалась основная проблема в теории теплового излучения и каков вклад Рэлея и Джинса в ее разрешение? 

Основная проблема в теории теплового излучения состояла в изучение спектрального состава излучения абсолютного черного тела.

Рэлей и Джинс представили равновесное излучение в полости как термодинамическую систему, обладающую определенным числом степеней свободы. Рэлей и Джинс применили к равновесному излучению в полости теорему классической статистической механики о равномерном распределении энергии по степеням свободы. В состоянии статистического равновесия на каждую колебательную степень свободы приходится средняя энергия, равная $kT$. Равновесное излучение в полости представляет собой систему стоячих волн. Число степеней свободы равно числу стоячих волн в полости.

Формула Рэлея-Джинса $r_{\nu, T}={{2\pi \nu^2}\over c^2}kT$ хорошо согласуется с экспериментальными данными в области больших длин волн, поэтому она с успехом используется в длинноволновой инфракрасной области спектра и в радиодиапазоне, но резко расходится с опытом для малых длин волн.

## Предпроцессинг данных и подключение библиотек

Код подключает необходимые библиотеки и проводит обработку данных из таблиц с данными, полученными при эксперименте.

```python
from csv import *
import pandas as pd
import numpy as np

def file_read(path):
    with open(path, 'r') as file:
        data = file.readlines()
    data[0] = data[0][3:]
    #print(data)
    #data = data[1:]
    for i in range(len(data)):
        data[i] = data[i].replace(',','.')
        data[i] = data[i].split(';')
        for j in range(len(data[i])):
            data[i][j] = float(data[i][j])
    #print(data)
    return data
```

## Вычисление постоянной Стефана-Больцмана

Из графика зависимости напряжения термостолбика от мощности электрического замещения видно, что $W_{зам} = 0,15U_{тс}$

Считывание данных из файла:

```python
table = file_read("45.csv")
table = np.array(table)
table = table.transpose()
table = table.tolist()
print(table)
```

Считаем W_зам

```python
W = []
for i in range(len(table[0])):
    W.append(round(table[1][i]*0.15,5))
table.append(W)
print(table[2])
```

Считаем T

```python
T = []
for i in range(len(table[0])):
    T.append(table[0][i]+273)
table.append(T)
print(table[3])
```

Считаем постоянную для каждой Т

```python
sigma = []
S = 1.12/(10**5)
for i in range(len(table[0])):
    sigma.append(round(table[2][i]/S/(table[3][i]**4),13))
table.append(sigma)
print(table[4])
```

```python
lines1 = {"t, (°C)":table[0],\
         "U_тс, (мВ)":table[1],\
         "W_зам, (Вт)":table[2],\
         "T, (K)":table[3],\
         "$\sigma$=W/(ST^4), (Дж/(м^2сK^4))":table[4]}
t1 = pd.DataFrame(lines1)
t1
```

Среднее значение постоянной:

```python
sigma_aver = sum(table[4])/len(table[4])
print(sigma_aver)
```

Оценка погрешности измерения:

```python
p_SB = 5.7*10**(-8)
diff = p_SB - sigma_aver
print(diff,",",diff/p_SB*100,"%")
```
